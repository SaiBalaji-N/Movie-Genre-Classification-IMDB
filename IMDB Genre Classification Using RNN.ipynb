```python
import pandas as pd
file_path="C:/Users/bsai3/Downloads/Genre Classification Dataset/train_data.txt"
with open(file_path,'r',encoding='utf-8') as file:
    data=file.readlines()
```

```python
records=[]
for line in data:
    record=line.strip().split(":::")
    records.append(record)
records[:5]
```

```python
columns=['s.no','movie_name','genre','description']
df=pd.DataFrame(records,columns=columns)
df.head()
```

```python
df=df.drop(labels=["s.no"],axis=1)
df.head()
```

```python
df["genre"].unique()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
labels={}
num=0
for i in df.genre.unique():
    if i not in labels:
        labels[i]=num
        num=num+1
labels
```

```python
def replace_labels(target):
    return labels[target]
df["target"]=df["genre"].map(replace_labels)
df.head()
```

```python
x=df["description"]
y=df["target"]
x,y
```

```python
len(x.unique())
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding,SimpleRNN,LSTM,Dense
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus_train=[]
for i in range(0,len(x)):
        review=re.sub('[^a-zA-Z]',' ',x[i])
        review=review.lower()
        review=review.split()
        review=[ps.stem(word)for word in review if not word in stopwords.words('english')]
        review=' '.join(review)
        corpus_train.append(review)
```

```python
corpus_train
```

```python
corpus_train=pd.Series(corpus_train)
len(corpus_train.unique())
```

```python
vocab_size=54075
```

```python
onehot=[one_hot(words,vocab_size)for words in corpus_train] 
onehot
```

```python
nltk.download('punkt')
max_words=max(len(nltk.word_tokenize(review)) for review in corpus_train)
max_words
```

```python
sent_length=1615
emb_vec=pad_sequences(onehot,padding='pre',maxlen=sent_length)
print(emb_vec)
```

```python
embedding_vector_features=100
RNN=Sequential()
RNN.add(Embedding(input_dim=vocab_size,output_dim=embedding_vector_features,input_length=sent_length))
RNN.add(SimpleRNN(units=100))
RNN.add(Dense(units=27,activation='softmax'))
RNN.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(RNN.summary())
```

```python
RNN.fit(emb_vec,y,epochs=10,batch_size=32)
```

```python
file_path="C:/Users/bsai3/Downloads/Genre Classification Dataset/test_data.txt"
with open(file_path,'r',encoding='utf-8') as file:
    data_test=file.readlines()
    records_test=[]
    for line in data_test:
        record=line.strip().split(":::")
        records_test.append(record)
columns=['s.no','movie_name','description']
df_test=pd.DataFrame(records_test,columns=columns)
x_test=df_test["description"]
corpus_test=[]
for i in range(0,len(x_test)):
        review=re.sub('[^a-zA-Z]',' ',x_test[i])
        review=review.lower()
        review=review.split()
        review=[ps.stem(word)for word in review if not word in stopwords.words('english')]
        review=' '.join(review)
        corpus_test.append(review)
onehot=[one_hot(words,vocab_size)for words in corpus_test]
emb_vec_test=pad_sequences(onehot,padding='pre',maxlen=sent_length)
```

```python
predictions=RNN.predict(emb_vec_test)
```

```python
import numpy as np
predicted_labels=np.argmax(predictions,axis=1)
```
